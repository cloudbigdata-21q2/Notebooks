```python
# 특정 행의 레코드 추출
print(df.loc[3]) # 인덱스 접근
print(df.iloc[3]) # 행번호 접근
```

```python
# 마지막 행의 출력
print("Shape:", df.shape)
print(df.loc[df.shape[0] - 1])
print(df.iloc[-1]) # 음수 인덱스 -> iloc에서만 사용 가능
print(df.tail(n=1))
```

```python
# 복수 개의 행을 출력, loc, iloc메서드에 행 정보 리스트 전달
df.loc[[0, 99, 999]] # 인덱스 0, 99, 999 인 레코드 출력
```

```python
# 특정 조건의 행, 열을 추출
subset2 = df.loc[:, ['country', 'year', 'lifeExp']] # : <- 전체 행, 
# country, year, lifeExp 컬럼만 추출
subset2.head()
```

```python
# 전체 행 데이터 중, 0, 2, 5(마지막) 컬럼 데이터 추출
# subset3 = df.iloc[:, [0, 2, 5]]
subset3 = df.iloc[:, [0, 2, -1]] # 음수 인덱스 접근
subset3.head()
```

```python
# range를 이용한 데이터의 추출
r = range(4) # 0 ~ 3
df.iloc[:, r].head() # 전체 행 중, 0 ~ 3 컬럼 추출
```

```python
# 데이터프레임을 년도별로 그룹화
grouped_year_df = df.groupby('year') 
type(grouped_year_df) # DataFrameGroupBy -> 그루핑된 정보를 담고 있는 객체
```

```python
# 그룹화된 데이터프레임의 산술평균
grouped_year_df['lifeExp'].mean() # 단일 컬럼의 평균
# grouped_year_df[['lifeExp', 'pop']].mean() # 복수 컬럼의 평균
```

```python
# 데이터 프레임에서 활용 가능한 기초 통계량들
grouped_year_df.describe()
```

```python
# 그룹화된 데이터의 빈도(갯수) 카운트
df.groupby('continent')['country'].nunique() # 유니크한 값의 갯수
```

```python
# 시각화
# 연도별 그룹화된 데이터 프레임에서 lifeExp의 평균 확인
life_exp = grouped_year_df['lifeExp'].mean()
life_exp
```
